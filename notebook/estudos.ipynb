```python
# especificar a versÃ£o da API

access_token = '<PEGAR O TOKEN DENTRO DE SETTINGS / Developer Settings / Personal access tokens >'
headers = {'Authorization' : 'Bearer '+ access_token,
           'X-GitHub-Api-Version':'2022-11-28'}

username = 'amzn'
url = f"https://api.github.com/users/{username}/followers"

list_followers = [] 
count = 1 

while True:
    try:
        url_page = f'{url}?page={str(count)}'
        response = requests.get(url_page, headers=headers)
        follower = response.json()

        if len(follower) > 0 :
            list_followers.append(follower) 
            count += 1
        else:
            break    
        
    except:
        break

len(list_followers)
len(list_followers[0])
```

## Desafio Pandas 

```python
## Padronizando a lista 
list_logins_followers = [] 

for page in list_followers:
    for follower in page:
        list_logins_followers.append(follower['login'])
                                     
len(list_logins_followers)                                    

```

```python
import pandas as pd 

df = pd.DataFrame()

df['Logins'] = list_logins_followers

df.to_csv('followers_list')
```
